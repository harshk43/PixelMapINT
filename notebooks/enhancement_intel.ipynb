```python
def calculate_coverage(mask, class_id):

    total_pixels = mask.size
    class_pixels = np.sum(mask == class_id)

    return (class_pixels / total_pixels) * 100
```

```python
change_report = {}

for class_id, class_name in CLASS_NAMES.items():

    old_cov = calculate_coverage(pred_mask_old, class_id)
    new_cov = calculate_coverage(pred_mask_new, class_id)

    difference = new_cov - old_cov

    change_report[class_name] = {
        "before_percent": round(old_cov, 2),
        "after_percent": round(new_cov, 2),
        "change_percent": round(difference, 2)
    }

change_report
```

```python
change_alerts = []

forest_change = change_report["forest_land"]["change_percent"]
urban_change = change_report["urban_land"]["change_percent"]

if forest_change < -10:
    change_alerts.append(
        "Significant green cover reduction detected."
    )

if urban_change > 10:
    change_alerts.append(
        "Rapid urban expansion / possible encroachment detected."
    )

print("Change Detection Alerts:\n")

for alert in change_alerts:
    print("-", alert)
```
